```python
def frequencytable(dataset,columnName):
    freq=pd.DataFrame(columns=["unique_value","frequency","Relative_freq","cumsum"])
    freq["unique_value"]=dataset[columnName].value_counts().sort_index().index
    freq["frequency"]=dataset[columnName].value_counts().sort_index().values
    freq["Relative_freq"]=freq["frequency"]/sum(freq["frequency"])*100
    freq["cumsum"]=freq["Relative_freq"].cumsum()
    return freq
```

```python
frequencytable(dataset,"ssc_p")
```

```python
frequencytable(dataset,"hsc_p")
```

```python
frequencytable(dataset,"mba_p")
```
